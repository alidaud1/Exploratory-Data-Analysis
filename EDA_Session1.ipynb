```python
# 2nd method for finding missing values (HEAT MAP)
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# reading country.xls file
df_country = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Data Sets/Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
# check for missing values
df_country.isnull().sum()
```

```python
df.columns
```

```python
# combine two datasets on the basis of Country Code column which is like PK
final_df = pd.merge(df,df_country,on='Country Code',how='left')
# how means left join , right join or inner join
final_df.head()
```

Observations

```python
# Observations-1 (What kind of observation u come up with this?)
final_df.Country.value_counts()
```

```python
# Doing this for creating PIE CHART
country_names= final_df.Country.value_counts().index
```

```python
country_val= final_df.Country.value_counts().values
```

```python
# PIE CHART
plt.pie(country_val,labels=country_names)
```

```python
# PIE CHART for top 3 countries using zomato
plt.pie(country_val[:3] ,labels=country_names[:3],autopct='%1.2f%%') # for percentage
```

My Observations
--> India's have most no of transactions then US and then UK

```python
# Observation-2
final_df.columns
```

```python
# Creating group by object
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
# aggregate rating
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
# converting in dataframe
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# renaming column 0 , creating variable ratings
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations
1. Ratings between 4.5-4.9 --> Excellent
2. Ratings between 4.0-4.4 --> Very Good
3. Ratings between 3.5-3.9 --> Good
4. Ratings between 2.5-3.4 --> Average
5. Ratings between 1.8-2.4 --> Poor
6. Ratings between 4.5-4.9 --> Not Rated

```python
ratings.head()
```

```python
# Visualization barplot
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y='Rating Count', data=ratings)
```

```python
# applying rating color feature
sns.barplot(x="Aggregate rating",y='Rating Count',hue='Rating color', data=ratings)
```

```python
# changing graph line color
sns.barplot(x="Aggregate rating",y='Rating Count',hue='Rating color', data=ratings,palette=['Purple','red','orange','green','yellow','blue'])
```

Observations
1. Not rated count is very high
2. Max number of ratings b/w 2.5 -- 3.4

---->Now if you have missing rating values you can add b/w 2.5 -- 3.4 as an average

```python
# count plot (use this for plotting with respect to categorical variable)
sns.countplot(x='Rating color',data=ratings,palette=['Purple','red','orange','green','yellow','blue'])
# frequency of rating color
```

**In-depth analysis or observation of dataset**

```python
# 1. find countries name that has given 0 ratings ?
# SQL query (SELECT * FROM Country Names WHERE [Aggregate rating] = 0.0;
final_df[final_df['Aggregate rating']==0.0].groupby('Country').size().reset_index()
```

Observation
1. Max no of 0 ratings ----> India

```python
# 2. find out which currency is used by which country ?
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# 3. Which countries do have online delieveries option?
final_df[final_df['Has Online delivery']=='Yes'].groupby('Country').size().reset_index()
```

```python
final_df[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```

Observation:
1. Online delivery option is availible in UAE and INDIA
